# nim_game
the game of nim goes in the way that either players(human and computer) pick objects from blue or red pile if the _
pile become empty on their turn will   win the game and remaining object will be consider as points(2 points for red and 3 point blue)
##types:
1:standad
2:mesere

```python
import random as rd #importing random library for selecting blue and red
blue_pile=15 #we have pile named as blue_pile
red_pile=15  #we have pile named as red_pile
count=0


def user(blue_pile,red_pile): #here we select manually from blue_pile and red_pile
    while True:
        select=input('select blue or red pile:') 
        if select=='blue':#blue is selected here
            selection='blue'
            break    
        elif select=='red':#red is selected here
            selection='red'
            break
        else:
            print('invalid input..')
    while True:
        try:
            move=int(input('user input your move from 1,2,3:'))
            if move in [1,2,3] and move<blue_pile and move<red_pile:
                return move,selection
            else:print('enter correct value from 1,2,3')
        except Exception as e:
            print(e)
            

def computer(blue_pile,red_pile): #the computer select from blue pile and red pile automaticallly
    pile_select=rd.randint(1,2) #computer generates 1 and 2 randomly to select red and blue pile
    if pile_select==1 :
        print('computer select blue:')
        mod_blue_pile=blue_pile%4  #computer divides both piles with multiples of 4 to form minimax algorithum
        if mod_blue_pile==0:
            return 1,3       #it will return 1 if there is odd number of object and return 3 if even number of pile
        else:
            return 1,mod_blue_pile
    if pile_select==2:
        print('computer select red ')
        mod_red_pile=red_pile%4  
        if mod_red_pile==0:
            return 2,3
        else:
            return 2,mod_red_pile
while True:
    mov,select=user(blue_pile,red_pile)
    if select=='red':
        red_pile-=mov
        total_score=count
        print(f'player chose{select} and remove:{mov} and pile updated to:{red_pile}')
        if red_pile<=1:
            print('player wins!!!')
            print(f'total_point earned is:{2*(-total_score+blue_pile+red_pile)}')
            break
    elif select=='blue':
        blue_pile-=mov
        print(f'player chose:{select} and remove:{mov} and pile updated to:{blue_pile}')
        if blue_pile<=1:
            print('player wins!!!')
            total_score=count
            print(f'total_point earned is:{3*(-total_score+blue_pile+red_pile)}') 
            break

    select_pile,move1=computer(blue_pile,red_pile)
    if select_pile==1 and move1<blue_pile:
        blue_pile-=move1
        count+=1
        print(f'computer remove:{move1} and updated blue pile is:{blue_pile}')
        if blue_pile<=1:
            print('computer wins!')
            total_score=count
            print(f'total_point earned is:{3*(-total_score+blue_pile+red_pile)}')
            break
    elif select_pile==2 and move1<red_pile:
        red_pile-=move1
        count+=1
        print(f'computer remove:{move1} and updated blue pile is:{red_pile}')
        if red_pile<=1:
           
            print('computer wins!')
            print(f'total_point earned is:{2*(-total_score+blue_pile+red_pile)}')
            break

print('game over')  #game will over if either player computer or human will win   
```
