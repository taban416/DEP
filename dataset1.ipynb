```python
import matplotlib.pyplot as mpy
mpy.figure(figsize=(18,4))
mpy.title('student_marks_graph',fontdict={'fontsize':23})
mpy.plot(data['name'],data['marks'])
mpy.xlabel('name')
mpy.ylabel('marks')
mpy.grid()
mpy.show()
```

```python
mpy.figure(figsize=(18,4))
mpy.title('student_city_graph',fontdict={'fontsize':23})
mpy.plot(data['name'],data['city'])
mpy.xlabel('name',fontdict={'fontsize':20})
mpy.ylabel('city',fontdict={'fontsize':20})
mpy.grid()
mpy.show()
```

```python
mpy.figure(figsize=(18,4))
mpy.title('student_height_graph',fontdict={'fontsize':23})
mpy.plot(data['name'],data['height'])
mpy.xlabel('name',fontdict={'fontsize':23})
mpy.ylabel('height',fontdict={'fontsize':23})
mpy.grid()
mpy.show()
```

```python
mpy.figure(figsize=(18,4))
mpy.title('student_grade_graph',fontdict={'fontsize':23})
mpy.plot(data['name'],data['grade'])
mpy.xlabel('name',fontdict={'fontsize':23})
mpy.ylabel('grade',fontdict={'fontsize':23})
mpy.grid()
mpy.show()
```
